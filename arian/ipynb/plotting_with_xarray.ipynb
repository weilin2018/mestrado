```python
import xarray as xr
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# import netcdf file
ncin = xr.open_dataset('/home/danilo/mestrado/arian/pcse_only.nc',decode_times=False)

# extract only temperature
temp = ncin.TEMP[0,:,:,:]

# printing dimensions to better understanding
print(temp.coords.dims)
```

```python
# plotting SST -> select only depth = 0
temp.isel(DEPTH=0).plot()
```

```python
# plotting near bottom tempertura
temp.isel(DEPTH=3).plot()
```
